```python
import pandas as pd

def my_func(user_id):
    return {"score": user_id * 2, "type": user_id + 12}

N = 8
realz = 1

results = []
for n in range(N):
    for r in range(realz):
        out = my_func(n)
        ordered = {'realization': r, 'user_id': n, **out}
        results.append(ordered)

df = pd.DataFrame(results)
df['extra'] = df['type'] + np.random.normal(0, 1, df.shape[0])
a = df['type'].values

df['channel'] = None
M = 2
for u in df["user_id"]:
    vec = np.random.randint(0,10,(M,1)) + 1j*np.random.randint(0,10,(M,1))  # complex vector

    # ensure we target exactly ONE row:
    idx = df.index[df["user_id"] == u].item()   # .item() -> exactly one match
    df.at[idx, "channel"] = vec
print(df)
# print(df['score'].values[3])

# results = [my_func(i) for i in range(N) for _ in range(realz)]
# df = pd.DataFrame(results)
# print(df)
# print(df.iloc[10, 0])
#
# results_dict = {(r, u): my_func(u) for r in range(realz) for u in range(N)}
# print(results_dict)
# print(results_dict[(0, 1)])
# print(results)
# print(results[3]['type'])
```

```python
import pandas as pd
import numpy as np

# Create a simple DataFrame with one empty column
df = pd.DataFrame({'ID': ['Item_1', 'Item_2']})
df['Data'] = None # Initialize the column

# Create a NumPy array
my_array = np.array([10, 20, 30.5])

# --- This is the key step ---
# Assign the ENTIRE array object to a SINGLE cell
df.at[0, 'Data'] = my_array

print("--- The DataFrame ---")
print(df)
print("\n" + "="*30 + "\n")

# Now, let's retrieve the object from that single cell and check its type
retrieved_object = df.at[0, 'Data']

print(f"The object in cell (0, 'Data') is: {retrieved_object}")
print(f"The type of the object is: {type(retrieved_object)}")
```
